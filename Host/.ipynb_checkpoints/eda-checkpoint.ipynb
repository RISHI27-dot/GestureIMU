```python
from typing import List

def kaiser_filter(y: List, n: int = 20, beta: int = 7):
    b = kaiser(n, beta=beta)
    a = 1
    yy = lfilter(b=b, a=a, x=y)
    return yy
```

```python
def get_average(col: pd.Series):
    resolution = 15
    col_len = len(col)
    # bat = col_len//resolution
    if len(col) != 105:
        add = pd.Series([col.mean() for _ in range(abs(len(col)-100))])
        col.append(add, ignore_index=True)
        # col.reset_index(drop=True)

    avg = []

    for i in range(0, col_len, resolution):
        avg.append(col.iloc[i:i+resolution].mean())
    return avg
len(get_average(zero[0]))
```

```python
def get_average_row(col: pd.Series):
    resolution = 10
    col_len = len(col)
    # bat = col_len//resolution
    if len(col) != 100:
        add = pd.Series([col.mean() for _ in range(abs(len(col)-100))])
        col.append(add, ignore_index=True)
        # col.reset_index(drop=True)

    avg = []

    for i in range(0, col_len, resolution):
        avg.append(col.iloc[i:i+resolution].mean())
    return avg
```

```python
import os

def generate_features(dir: str, label: int, mean: str, n_features: int):
    features = []
    target = []
    for csv in os.listdir(dir):
        l = int(csv.split('_')[0])
        if l>label:
            break

        scale = MinMaxScaler(feature_range=(-1,1))
        df = pd.read_csv(os.path.join(dir, csv))
        df = pd.DataFrame(scale.fit_transform(df))
        df.drop(labels=[1,2], inplace=True, axis=1)
        df = df.iloc[:100, :]
        if mean:
            ax_processed = [float(x) for x in get_average(df)]
        else:
            ax_processed = [float(x) for x in kaiser_filter(y=df[0])]
        features.append(ax_processed)
        target.append(l)
        # print(features)
    
    label_csv = pd.DataFrame(features, columns=[f'x{i}' for i in range(n_features)] )
    label_csv['target'] = target
    filename = 'average_filtered' if mean else 'kaiser_filtered'
    label_csv.to_csv(f'./normalised_features_{filename}.csv', index=False)
    display(label_csv)
```

```python
import warnings
warnings.filterwarnings(action='once')

dir = './complete_dataset/csv'
generate_features(dir, 9, mean=True, n_features=7)
```

```python
feats = pd.read_csv('./normalised_features_average_filtered.csv')
feats.fillna(feats.median(), inplace=True)
# feats = get_average(feats)
feats
```

```python
import seaborn as sns
sns.heatmap(feats.isna(), cbar=False)
```

```python
feats['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

X_train, X_val, y_train, y_val = train_test_split(feats.iloc[:,:7], feats['target'], test_size=0.17, random_state=42, shuffle=True)
clf1 = SVC(gamma='auto')
clf3 = KNeighborsClassifier()
clf1.fit(X_train, y_train)
clf3.fit(X_train, y_train)
preds1 = clf1.predict(X_val)
preds3 = clf3.predict(X_train)
display(accuracy_score(y_val, preds1))
display(accuracy_score(y_train, preds3))
```

```python
train_preds = clf3.predict(X_train)
val_preds = clf3.predict(X_val)

display(print(f'Train accuracy: {accuracy_score(y_train, train_preds)}'))
display(print(f'Validation accuracy: {accuracy_score(y_val, val_preds)}'))
```

```python
from keras.layers import Dense, Input
from keras.models import Model

inp = Input((7))
fc1 = Dense(32, activation='relu')(inp)
fc2 = Dense(16, activation='relu')(fc1)
fc3 = Dense(10, activation='softmax')(fc2)

model = Model(inputs=inp, outputs=[fc3])
model.build(input_shape=(7))
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
hist = model.fit(
    X_train,
    y_train,
    epochs=30,
    validation_data=(X_val, y_val),
    verbose=1,
    shuffle=True
)
```

```python
import tensorflow as tf

converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

with open('./nn_gesture.tflite', 'wb') as f:
    f.write(tflite_model)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cfm = confusion_matrix(y_train, clf3.predict(X_train), labels=[i+1 for i in range(9)])
sns.heatmap(cfm, cbar=False, annot=True, cmap='Blues')
plt.title('K Nearest Neighbors confusion matrix')
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

nn_preds = []
for pred in model.predict(X_train):
    pred = np.asarray(pred)
    idx = pred.argmax()
    nn_preds.append(idx)

cfm = confusion_matrix(y_train, nn_preds, labels=[i+1 for i in range(9)])
sns.heatmap(cfm, cbar=False, annot=True, cmap='Blues')
plt.title('Feed forward neural net confusion matrix')
```
